## Data Vizualization and Analysis 

```python
!pip install pyfolio
```

```python
ticker_list = ['TSLA', 'AAPL', 'MSFT', 'TWTR']

import pandas as pd 
data = pd.DataFrame(columns = ticker_list)

#get data
for ticker in ticker_list:
    data[ticker] = yf.download(ticker, period = '5y',)['Adj Close']
    
#compute the return of the stocvks and then compute the daily avg returns 
# the mean return is the daily portfolio returns with the above four stocks 

data = data.pct_change().dropna().mean(axis = 1)

import pyfolio as pf 

pf.create_simple_tear_sheet(data)
```
