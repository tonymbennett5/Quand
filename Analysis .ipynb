## Moving Averages 

```python
ap_data.head()
```

```python

ap_data['MA_9'] = ap_data['close'].rolling(9).mean()
ap_data['MA_21'] = ap_data['close'].rolling(21).mean()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,10))
plt.grid(True)
plt.plot(ap_data['close'], Label = "Close" )
plt.plot(ap_data['MA_9'], Label = "MA 9 Day")
plt.plot(ap_data['MA_21'], Label = "MA 21 Day")
plt.legend(loc=2)
```
